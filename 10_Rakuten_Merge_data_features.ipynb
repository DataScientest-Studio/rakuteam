```
# Referencer une df product
df_prod = X.productid

# Compter les doublons
df_prod.duplicated().sum()
```

```
# Merge feats_aylin
df_prod = df_prod.to_frame()
fusion_1 = df_prod.merge(right = feats_aylin, on='productid', how = 'left')
```

```
fusion_1.info()
```

```
# Merge feats_damien
fusion_2 = fusion_1.merge(right = feats_damien, on='productid', how = 'left')
```

```
fusion_2.info()
```

```
# Merge feats_julien

#cols_to_use = feats_julien.columns.difference(fusion_2.columns)
#cols_to_use = cols_to_use.to_frame()

fusion_3 = fusion_2.merge(right = feats_julien, on='productid', how = 'left')

#fusion_3 = merge(fusion_2, feats_julien[cols_to_use], left_index=True, right_index=True, how='left')
```

```
fusion_3.isnull().sum()
```

```
# Supprimer les colonnes identiques entre df Damien et Julien
to_drop = ['prdtypecode_y','designation_y','description_y', 'description_y', 'Unnamed: 0_y', 'Unnamed: 0.1']
data_features = fusion_3.drop(to_drop, axis = 1)
```

```
# Renommer les colonnes
names = {'prdtypecode_x': 'prdtypecode', 'Unnamed: 0_x' : 'Unnamed: 0', 'designation_x': 'designation', 'description_x': 'description'}
data_features.rename(columns = names, inplace = True)
```
